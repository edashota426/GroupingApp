```python
#inport pandas & numpy
import pandas as pd
import numpy as np

#import csv and  convert it to pandas dateframe
df = pd.read_csv("sample.csv", encoding="utf_8")
#add calum for "group"
df["group"] = int(100)
#sort the dateframe by "skill"
sorted_df = df.sort_values(by='jf').sort_values(by='presentation').sort_values(by='region').sort_values(by='skill')
#devide dateframe into "male" & "female"
male = sorted_df[sorted_df.gender == "M"]
female = sorted_df[sorted_df.gender == "F"]
#set group number
group_number = 4
#insert group number for each gender
male["group"]  = list(range(group_number))[::-1]*( int(len(male)/group_number)) + list(range(int(len(male)%group_number)))
female["group"]  = list(range(group_number))*( int(len(female)/group_number)) + list(range(group_number))[::-1][:len(female)%group_number]
#convert two gender groups
labeled_df = pd.concat([male, female])
#sort the result dataframe by "group"
sorted_labeled_df = labeled_df = labeled_df.sort_values(by="group")
sorted_labeled_df
```

```python
sorted_labeled_df.to_csv("newsample.csv", encoding="shift_jis")
```
